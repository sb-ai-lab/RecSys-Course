### Hit Rate

$
HitRate@k(u) =\mathbb{I}[|rel(u) \cap rec_k(u)| > 0]
$

$
HitRate@k =\frac 1N  \sum_{u=1}^N HitRate@k(u)$


```python
def hit_rate(test:pd.DataFrame, predict:pd.DataFrame, topk:int) -> float:
    hit_rate_list = []
    user_pred = set(predict['user_id'].unique())
    user_test = set(test['user_id'].unique())
    user_union = set.intersection(user_pred, user_test)
    predict.sort_values(['user_id', 'rating'], inplace=True, ascending=False)
    for user in user_union:
        test_seq = set(test[test['user_id'] == user]['item_id'].values)
        pred_seq = predict[predict['user_id'] == user]['item_id'].values
        if len(set(pred_seq[:topk]).intersection(test_seq)) > 0:
            hit_rate_list.append(1)
        else:
            hit_rate_list.append(0)

    no_test_users = user_test - user_union
    hit_rate_list.extend([0] * len(no_test_users))

    return np.mean(hit_rate_list)
```

### Precision

$Precision@k(u) = \frac{|rel(u) \cap rec_k(u)|}{k}$


$Precision@k =\frac 1N  \sum _{u=1}^N Precision@k(u)$


```python
def precision(test:pd.DataFrame, predict:pd.DataFrame, topk:int) -> float:
    res = []
    user_pred = set(predict['user_id'].unique())
    user_test = set(test['user_id'].unique())
    user_union = set.intersection(user_pred, user_test)
    predict.sort_values(['user_id', 'rating'], inplace=True, ascending=False)

    for user in user_union:
        test_seq = test[test['user_id'] == user]['item_id'].values
        pred_seq = predict[predict['user_id'] == user]['item_id'].values

        ans = len(set(pred_seq[:topk]) & set(test_seq))
        res.append(ans / topk)

    no_test_users = user_test - user_union
    res.extend([0] * len(no_test_users))

    return np.mean(res)
```

### Recall

$Recall@k(u) = \frac{|rel(u) \cap rec_k(u)|}{|rel(u)|}$


$Recall@k =\frac 1N  \sum_{u=1}^N Recall@k(u)$


```python
def recall(test: pd.DataFrame, predict : pd.DataFrame, topk:int) -> float:
    res = []
    user_pred = set(predict['user_id'].unique())
    user_test = set(test['user_id'].unique())
    user_union = set.intersection(user_pred, user_test)
    predict.sort_values(['user_id', 'rating'], inplace=True, ascending=False)

    for user in user_union:
        test_seq = test[test['user_id'] == user]['item_id'].values
        pred_seq = predict[predict['user_id'] == user]['item_id'].values

        ans = len(set(pred_seq[:topk]).intersection(set(test_seq)))
        res.append(ans / len(set(test_seq)))

    no_test_users = user_test - user_union
    res.extend([0] * len(no_test_users))

    return np.mean(res)
```

### NDCG

$CG@k(u) =  \sum_{i \in rec_k(u)} \mathbb{I}[i \in rel(u)]$ \
$DCG@k(u) =  \sum_{i \in rec_k(u)} \frac{\mathbb{I}[i \in rel(u)]} {\log_2(rank(u, i)+1)}$ \
$IDCG$ - максимально возможный $DCG$ для данного пользователя, когда все релавантные товары стоят вначале


$ NDCG@k(u) = \frac{DCG@k(u)} {IDCG@k(u)}$ \
$NDCG@k =\frac 1N  \sum_{u=1}^N NDCG@k(u)$



```python
def ndcg(test: pd.DataFrame, predict : pd.DataFrame, topk:int) -> float:
    res = []
    user_pred = set(predict['user_id'].unique())
    user_test = set(test['user_id'].unique())
    user_union = set.intersection(user_pred, user_test)
    predict.sort_values(['user_id', 'rating'], inplace=True, ascending=False)

    for user in user_union:
        test_seq = test[test['user_id'] == user]['item_id'].values
        pred_seq = predict[predict['user_id'] == user]['item_id'].values

        predict_len = min(topk, len(pred_seq))
        test_len = min(topk, len(test_seq))
        denom = [1 / np.log2(i + 2) for i in range(topk)]
        dcg = sum(denom[i] for i in range(predict_len) if pred_seq[i] in test_seq)
        idcg = sum(denom[:test_len])
        res.append(dcg / idcg)


    no_test_users = user_test - user_union
    res.extend([0] * len(no_test_users))

    return np.mean(res)
```

$NDCG@3(u)$ = $DCG@3(u)$ / $IDCG@3(u)$ = 0.63 / 2.13 ~ 0.3

### Coverage

$$Coverage@K=\frac{|\bigcup \limits_{u\in U} rec_k(u)|
}{|I|}$$

```python
def coverage(train:pd.DataFrame, predict:pd.DataFrame, topk:int) -> float:
    unique_item = set(train['item_id'])
    predict.sort_values(['user_id', 'rating'], inplace=True, ascending=False)
    predict_topk = predict.groupby('user_id', group_keys=True)['item_id'].apply(lambda x: x.head(topk)).reset_index()['item_id']
    unique_predict = set(predict_topk)
    return len(unique_predict & unique_item) / len(unique_item)
```

### Классы OfflineMetrics и experiments

Для подсчета множества метрик удобнее всего использовать модуль OfflineMetrics из Replay. Он позволяет считать различные метрики в разных режимах: PerUser, Median, Mean, ConfidenceInterval.

```python
from replay.metrics import Median, ConfidenceInterval, PerUser, MAP, MRR, NDCG, RocAuc, Novelty, Surprisal, OfflineMetrics
metrics = [
     Precision(topk=2),
     Precision(topk=5, mode=PerUser()),
     Precision(topk=10, mode=Median()),
     Precision(topk=5, mode=ConfidenceInterval(alpha=0.95)),
     MAP(topk=10),
     MRR(topk=20),
     NDCG(topk=10),
     HitRate(topk=5),
     RocAuc(topk=10),
     Coverage(topk=10),
     Novelty(topk=5),
     Surprisal(topk=10),
 ]

OfflineMetrics(metrics, query_column='user_id')(pred, test, train)
```
